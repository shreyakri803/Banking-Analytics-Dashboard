# Advances analaysis based on Dataset
###  Categorizing Estimated Income into Bands for Distribution Analysis

```python
bins= [0,100000, 300000, float('inf')]
labels = ['Low', 'Med' , 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins , labels=labels , right= False)
```

```python




income_counts = df['Income Band'].value_counts().sort_index()

# Plot
income_counts.plot(kind='bar', color=['#66c2a5', '#fc8d62', '#8da0cb'], figsize=(6, 4))

# Add labels and title
plt.title('Income Band Distribution')
plt.xlabel('Income Band')
plt.ylabel('Number of Customers')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

### ðŸ§¾ Step 3: Exploring Categorical Variables

In this step, we examine the distribution of unique categories in selected categorical columns such as:
- Amount of Credit Cards
- Nationality
- Occupation
- Fee Structure
- Loyalty Classification
- Properties Owned
- Risk Weighting
- Income Band

```python
# List of all categorical columns for analysis
categorical_cols = df[[ 
    "Amount of Credit Cards", 
    "Nationality", 
    "Occupation", 
    "Fee Structure", 
    "Loyalty Classification", 
    "Properties Owned", 
    "Risk Weighting", 
    "Income Band"
]].columns

# Display value counts for each categorical column
for col in categorical_cols:
    print(f"\nValue Counts for '{col}':")
    display(df[col].value_counts())
```

### ðŸ“Š Univariate Analysis of Categorical Features

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Loop through each categorical variable and create a count plot
for i, predictor in enumerate(df[[
    "BRId", 
    "GenderId", 
    "IAId", 
    "Amount of Credit Cards", 
    "Nationality", 
    "Occupation", 
    "Fee Structure", 
    "Loyalty Classification", 
    "Properties Owned", 
    "Risk Weighting", 
    "Income Band"
]]):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
    plt.title(f"Distribution of {predictor}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

### ðŸ“Š Biivariate Analysis of Categorical Features


```python
for i, predictor in enumerate(df[[
    "BRId", 
    "IAId", 
    "Amount of Credit Cards", 
    "Nationality", 
    "Occupation", 
    "Fee Structure", 
    "Loyalty Classification", 
    "Properties Owned", 
    "Risk Weighting", 
    "Income Band"
]]):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='GenderId')
    plt.title(f"{predictor} Distribution by Gender")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Bivariate analysis: categorical features vs Nationality
for i, predictor in enumerate(df[[
    "BRId", 
    "IAId", 
    "Amount of Credit Cards", 
    "GenderId", 
    "Occupation", 
    "Fee Structure", 
    "Loyalty Classification", 
    "Properties Owned", 
    "Risk Weighting", 
    "Income Band"
]]):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='Nationality')
    plt.title(f"{predictor} Distribution by Nationality")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

### ðŸ“Š Histogram of Occupation Distribution

In this step, we visualize the frequency of different occupation categories using a histogram (countplot). This helps us understand which occupations are most common among customers and identify any imbalance or underrepresented groups in the data.


```python
# Histograms of value counts for all categorical columns
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col])
    plt.title(f"Histogram of {col} Count")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

```

## Numerical Analysis 

```python
numerical_cols = [
    'Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 
    'Bank Loans', 'Bank Deposits', 'Checking Accounts', 
    'Saving Accounts', 'Foreign Currency Account', 'Business Lending'
]

# Univariate analysis and visualization
plt.figure(figsize=(15, 22))  # Adjusted height to fit all plots
for i, col in enumerate(numerical_cols):
    plt.subplot(len(numerical_cols), 1, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
```

## Heatmaps

```python
numerical_cols = [
    'Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 
    'Bank Loans', 'Bank Deposits', 'Checking Accounts', 
    'Saving Accounts', 'Foreign Currency Account', 'Business Lending'
]

# Calculate correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Create the heatmap
plt.figure(figsize=(12, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

## ðŸ” Exploratory Data Analysis (EDA) Insights â€“ Correlation Heatmap

### ðŸ§¾ Overview
This heatmap displays the Pearson correlation between key financial metrics of banking customers. The objective is to identify patterns in financial behavior, product relationships, and potential opportunities for customer targeting or risk analysis.

---

### ðŸ“Š Key Correlation Insights

#### ðŸ’¡ Strong Positive Correlations:
- **Bank Deposits â†” Checking Accounts (0.84)**  
  Customers with high deposits often maintain high checking balances, suggesting a strong engagement with liquid banking products.

- **Bank Deposits â†” Saving Accounts (0.75)**  
  Indicates that individuals who deposit more also save more, presenting an opportunity for long-term investment products.

- **Checking Accounts â†” Saving Accounts (0.46)**  
  Moderate correlation suggesting that many customers utilize both accounts, reflecting healthy financial behavior and cross-product engagement.

#### ðŸ’¡ Moderate Relationships:
- **Business Lending â†” Bank Deposits (0.44)**  
  Business loan customers tend to have higher deposits, possibly due to managing business cash flow.

- **Business Lending â†” Bank Loans (0.42)**  
  Clients using business lending products are likely to use personal or additional credit products as well.

- **Credit Card Balance â†” Bank Deposits (0.38)**  
  A moderate correlation that may indicate credit use backed by available funds or higher spending ability.

#### âš ï¸ Weak or Low Correlations:
- **Superannuation Savings** and other daily banking metrics show weak correlations. This is expected, as superannuation tends to be influenced by external factors like employment type or long-term goals rather than short-term banking behavior.

---

### ðŸ’¼ Business Value Takeaways
- **High deposit + checking balance customers** may be ideal candidates for premium or advisory banking services.
- The linkage between **lending products** (Bank Loans & Business Lending) opens up opportunities for bundled financial solutions.
- Customers showing consistent use of multiple accounts can be prioritized for **retention and loyalty** strategies.
